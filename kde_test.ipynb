```python
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV
import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(3, 0.5, size=(99,1))
```

```python
params = {'bandwidth': np.logspace(-1, 1, 20)}
grid = GridSearchCV(KernelDensity(kernel='gaussian'), params)
grid.fit(data)

kde = grid.best_estimator_
```

```python
fig, ax = plt.subplots()
ax.scatter(data.flatten(),np.zeros_like(data.flatten()))
xs = np.linspace(np.min(data), np.max(data)).reshape(-1,1)
ys = np.exp(kde.score_samples(xs))
ax.plot(xs, ys)
plt.show()
```

```python
from scipy.stats import differential_entropy
differential_entropy(data)
```
